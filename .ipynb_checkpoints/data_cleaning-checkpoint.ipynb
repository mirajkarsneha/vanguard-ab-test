```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Client Profiles dataset
df_final_demo = pd.read_csv('/Users/rishikeshdhokare/Documents/Ironhack/MiniProject/project2-vanguard/DataFiles/final_data.csv')

# 1. Age Distribution
sns.histplot(df_final_demo['clnt_age'], bins=20, kde=True)
plt.show()

# Grouping ages into bins to identify the most common age group
age_bins = [18, 25, 35, 45, 55, 65, 75, 85, 95]
df_final_demo['age_group'] = pd.cut(df_final_demo['clnt_age'], bins=age_bins, labels=[
    '18-24', '25-34', '35-44', '45-54', '55-64', '65-74', '75-84', '85+'])

# Plot the age group distribution
sns.countplot(data=df_final_demo, x='age_group', order=['18-24', '25-34', '35-44', '45-54', '55-64', '65-74', '75-84', '85+'])
plt.show()

# 2. Client Tenure Analysis
df_final_demo['tenure_group'] = pd.cut(df_final_demo['clnt_tenure_yr'], bins=[0, 1, 5, 10, 20, 50], labels=[
    'Less than 1 year', '1-5 years', '5-10 years', '10-20 years', '20+ years'])

# Plot the tenure group distribution
sns.countplot(data=df_final_demo, x='tenure_group')
plt.show()

# 3. Gender Distribution
sns.countplot(data=df_final_demo, x='gender')
plt.show()

# Display the summary of the primary client characteristics
primary_clients_summary = df_final_demo.groupby(['clnt_age', 'clnt_tenure_yr', 'gender']).size().reset_index(name='client_count')
print(primary_clients_summary)
```

```python
# Get the most common age group
most_common_age_group = df_final_demo['age_group'].mode()[0]

# Get the most common tenure group
most_common_tenure_group = df_final_demo['tenure_group'].mode()[0]

# Get the most common gender
most_common_gender = df_final_demo['gendr'].mode()[0]

# Print the results
print(f"The primary clients are predominantly aged in the '{most_common_age_group}' group, "
      f"with '{most_common_tenure_group}' tenure, and are mostly '{most_common_gender}'.")
```

```python
most_common_age_group = df_final_demo['age_group'].mode()[0]
if most_common_age_group in ['18-24', '25-34', '35-44']:
    age_category = 'Younger'
else:
    age_category = 'Older'

most_common_tenure_group = df_final_demo['tenure_group'].mode()[0]
if most_common_tenure_group in ['Less than 1 year', '1-5 years']:
    tenure_category = 'New'
else:
    tenure_category = 'Long-standing'

print(f"The primary clients using the online process are {age_category.lower()} and {tenure_category.lower()} clients.")
```

```python
import pandas as pd
import numpy as np

# Load your data
df_final_experiment_clients = pd.read_csv('/Users/rishikeshdhokare/Documents/Ironhack/MiniProject/project2-vanguard/DataFiles/final_data.csv')

# Clean the 'Variation' column
df_final_experiment_clients['Variation'] = df_final_experiment_clients['Variation'].str.strip().str.lower()

# Calculate the total number of unique clients who started the process
clients_started = df_final_experiment_clients['client_id'].nunique()

# Calculate the total number of unique clients who reached "confirm"
clients_reached_confirm = df_final_experiment_clients[
    df_final_experiment_clients['process_step'] == 'confirm'
]['client_id'].nunique()

# Calculate the overall completion rate
overall_completion_rate = (clients_reached_confirm / clients_started) * 100
print(f"Overall Completion Rate: {overall_completion_rate:.2f}%")

# Calculate completion rates for test and control groups
test_group_clients = df_final_experiment_clients[df_final_experiment_clients['Variation'] == 'test']
control_group_clients = df_final_experiment_clients[df_final_experiment_clients['Variation'] == 'control']

test_group_completion = test_group_clients[test_group_clients['process_step'] == 'confirm']['client_id'].nunique()
control_group_completion = control_group_clients[control_group_clients['process_step'] == 'confirm']['client_id'].nunique()

test_completion_rate = (test_group_completion / test_group_clients['client_id'].nunique()) * 100
control_completion_rate = (control_group_completion / control_group_clients['client_id'].nunique()) * 100

print(f"Test Group Completion Rate: {test_completion_rate:.2f}%")
print(f"Control Group Completion Rate: {control_completion_rate:.2f}%")
```

```python
# Part 2: Analyzing the Steps Leading Up to "Confirm"
# Define the required steps in order
required_steps = ['start', 'step_1', 'step_2', 'step_3', 'confirm']

# Group by 'client_id' and aggregate the steps each client completed
steps_completed_per_client = df_final_experiment_clients.groupby('client_id')['process_step'].unique()

# Calculate how many clients completed all the required steps in order
clients_completed_all_steps = steps_completed_per_client.apply(lambda x: all(step in x for step in required_steps)).sum()

# Calculate the rate of clients completing all steps
completion_rate_all_steps = (clients_completed_all_steps / clients_started) * 100
print(f"Completion Rate of All Steps: {completion_rate_all_steps:.2f}%")
```
